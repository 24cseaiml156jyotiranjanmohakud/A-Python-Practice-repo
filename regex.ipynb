```python
import regex as re
```

```python
pattern="25"
text="Jyoti's birthday is on 04-09-2006"

match=re.search(pattern,text)
print(match)
print(type(match))
```

```python
pattern="627"
text="My Roll no is: 24CSEAIML156"

match=re.search(pattern,text)
print(match)
print(type(match))
```

```python
string="Hello I am Jyoti ! your friend 156"
pattern="\d+"

result=re.findall(pattern,string)
print(result)
```

```python
string="Hello I am Jyoti ! your friend 156 hello 131"
pattern="\d+"

result=re.finditer(pattern,string)
print(result)
```

```python
pattern="Jyoti"
text="Jyoti's birthday is on 04-09-2006"

match=re.search(pattern,text,re.IGNORECASE)
print(match)
print(type(match))
```
